```python
import numpy as np
import pandas as pd


# Convert data into a DataFrame
df = pd.DataFrame(df)

# Filter out only numerical columns
numerical_cols = df.select_dtypes(include=[np.number])

# Calculate the first quartile (Q1) for each numerical column
q1 = numerical_cols.quantile(0.25)

# Calculate the third quartile (Q3) for each numerical column
q3 = numerical_cols.quantile(0.75)

# Calculate the interquartile range (IQR) for each numerical column
iqr = q3 - q1

print("First Quartile (Q1):")
print(q1)
print("\nThird Quartile (Q3):")
print(q3)
print("\nInterquartile Range (IQR):")
print(iqr)
```

```python
# Define the lower and upper bounds for outlier detection
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Identify outliers for each numerical column
outliers = (numerical_cols < lower_bound) | (numerical_cols > upper_bound)

# Count the number of outliers in each numerical column
num_outliers = outliers.sum()

print("Number of outliers in each numerical column:")
print(num_outliers)

# Remove rows containing outliers
df_data= df[~outliers.any(axis=1)]

print("\nDataFrame after removing outliers:")
df_data
```

```python
dates = pd.date_range(start='2011-01-01', end='2014-12-31')
df = pd.DataFrame({'Date': dates, 'Data': np.random.randint(0, 2, len(dates))})

# Function to check if a given date is a weekday
def is_weekday(date):
    return date.weekday() < 5

# Function to check if a given date is a holiday
def is_holiday(date):
    holidays = [(1, 1), (12, 25)]  # New Year's Day, Independence Day, Christmas Day
    return (date.month, date.day) in holidays

# Process data for each year
years_data = {}
for year in range(2011, 2015):
    year_mask = (df['Date'].dt.year == year)
    year_df = df[year_mask].copy()  # Extract data for the current year

    # Check weekdays and holidays
    year_df['Weekday'] = year_df['Date'].apply(lambda x: 1 if is_weekday(x) else 0)
    year_df['Holiday'] = year_df.apply(lambda row: 0 if row['Weekday'] == 1 and is_holiday(row['Date']) == 0 else 1, axis=1)

    # Store processed data for the year
    years_data[year] = year_df

# Example usage: Access data for a specific year
print("Data for 2011:")
print(years_data)


```

```python
all_years_data = pd.concat(years_data.values(), ignore_index=True)
all_years_data
```

```python
X = merged_df.drop(columns=['Profit'],axis=1)
y = merged_df['Profit']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=24)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
```

```python
tree = DecisionTreeRegressor(max_depth=4,max_features=4)

tree.fit(X_train, y_train)
```

```python
predictions = tree.predict(X_test)
```

```python
df=pd.DataFrame({'Actual':y_test, 'Predicted':predictions})
df.head(5)
```

```python
# Evaluating the Prediction Accuracy
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test,predictions))
print('Mean Squared Error:', metrics.mean_squared_error(y_test,predictions))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test,predictions)))
print('r2_score:', metrics.r2_score(y_test,predictions))
```
